```python
import numpy as np
```

```python
a = np.array([[1, 6],[2, 8], [3, 11], [3, 10], [1, 7]])
```

```python
a.shape
```

```python
mean_a = np.array([np.mean(a[:,0]), np.mean(a[:,1])])
```

```python
mean_a
```

```python
a_centered = np.subtract(a, mean_a)
```

```python
a_centered
```

```python
a_centered_sp = np.dot(a_centered[:,0], a_centered[:,1])
```

```python
a_centered_sp / (np.shape(a)[1] - 1)
```
